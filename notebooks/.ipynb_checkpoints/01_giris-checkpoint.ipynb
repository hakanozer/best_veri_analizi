# Giriş
Bu not defteri veri analizine giriş içindir.

```python
# Hücre 1 — Kütüphaneleri kontrol et
import sys
print(f"Python versiyonu: {sys.version}")

# Hücre 2 — Basit hesap
fiyat = 150.0
adet = 3
toplam = fiyat * adet
print(f"Toplam tutar: {toplam} TL")
```

```python
# Değişken atama — tip bildirmenize gerek yok
musteri_adi = "Ahmet Yılmaz"   # str
siparis_tutari = 250.75         # float
siparis_adedi = 5               # int
uye_mi = True                   # bool

# Tip kontrolü
print(type(musteri_adi))        # <class 'str'>
print(type(siparis_tutari))     # <class 'float'>

# F-string (modern string biçimlendirme)
print(f"{musteri_adi} müşterisinin toplam siparişi: {siparis_tutari} TL")
```

```python
# if / elif / else
siparis_tutari = 350

if siparis_tutari >= 500:
    print("Ücretsiz kargo hakkı kazandınız!")
elif siparis_tutari >= 200:
    print("50 TL kargo ücreti.")
else:
    print("Standart kargo ücreti uygulanır.")

# for döngüsü
urunler = ["Laptop", "Mouse", "Klavye"]
for urun in urunler:
    print(f"Ürün: {urun}")

# while döngüsü
sayac = 0
while sayac < 3:
    print(f"Deneme: {sayac}")
    sayac += 1
```
