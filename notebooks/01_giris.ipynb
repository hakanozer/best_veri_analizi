```python
# if / elif / else
siparis_tutari = 350

if siparis_tutari >= 500:
    print("Ücretsiz kargo hakkı kazandınız!")
elif siparis_tutari >= 200:
    print("50 TL kargo ücreti.")
else:
    print("Standart kargo ücreti uygulanır.")

# for döngüsü
urunler = ["Laptop", "Mouse", "Klavye"]
for urun in urunler:
    print(f"Ürün: {urun}")

# while döngüsü
sayac = 0
while sayac < 3:
    print(f"Deneme: {sayac}")
    sayac += 1
```

```python
sehirler = ["İstanbul", "Ankara", "İzmir", "Bursa", "Antalya", "Adana", "Konya", "Gaziantep", "Mersin", "Kayseri"]
print(sehirler)
print("\nList için en çok kullanılan fonksiyonlar ve örnekleri:\n")

# 1) append()
liste = sehirler.copy()
liste.append("Eskişehir")
print("1. append(x): Listenin sonuna eleman ekler.")
print("Sonuç:", liste, "\n")

# 2) insert()
liste = sehirler.copy()
liste.insert(2, "Eskişehir")
print("2. insert(i, x): Belirtilen indekse eleman ekler.")
print("Sonuç:", liste, "\n")

# 3) extend()
liste = sehirler.copy()
liste.extend(["Trabzon", "Samsun"])
print("3. extend(iterable): Listeye birden fazla eleman ekler.")
print("Sonuç:", liste, "\n")

# 4) remove()
liste = sehirler.copy()
liste.remove("Bursa")
print("4. remove(x): Verilen değere sahip ilk elemanı siler.")
print("Sonuç:", liste, "\n")

# 5) pop()
liste = sehirler.copy()
silinen = liste.pop()
print("5. pop([i]): Belirtilen indeksteki elemanı siler ve geri döndürür.")
print("Silinen eleman:", silinen)
print("Sonuç:", liste, "\n")

# 6) index()
liste = sehirler.copy()
konum = liste.index("İzmir")
print("6. index(x): Elemanın listedeki ilk indeksini verir.")
print("'İzmir' indeksi:", konum, "\n")

# 7) count()
liste = ["Ankara", "İzmir", "Ankara", "Bursa", "Ankara"]
adet = liste.count("Ankara")
print("7. count(x): Bir elemanın listede kaç kez geçtiğini sayar.")
print("'Ankara' sayısı:", adet, "\n")

# 8) sort()
liste = sehirler.copy()
liste.sort()
print("8. sort(): Listeyi kalıcı olarak sıralar.")
print("Sonuç:", liste, "\n")

# 9) reverse()
liste = sehirler.copy()
liste.reverse()
print("9. reverse(): Listenin sırasını ters çevirir.")
print("Sonuç:", liste, "\n")

# 10) copy()
liste = sehirler.copy()
print("10. copy(): Listenin kopyasını oluşturur.")
print("Kopya liste:", liste)
print("Aynı nesne mi?:", liste is sehirler, "\n")

# 11) clear()
liste = sehirler.copy()
liste.clear()
print("11. clear(): Listedeki tüm elemanları siler.")
print("Sonuç:", liste, "\n")

# Ek olarak sık kullanılan yerleşik fonksiyonlar
print("Ek bilgiler:")
print("len(liste): Listedeki eleman sayısını verir ->", len(sehirler))
print("max([3, 7, 2, 9]): En büyük değeri verir ->", max([3, 7, 2, 9]))
print("min([3, 7, 2, 9]): En küçük değeri verir ->", min([3, 7, 2, 9]))

sehirler.sort(reverse = True)
print("Sıralanmış şehirler:", sehirler)
```

```python
karmasik_veri = {
    "musteri": {
        "ad": "Ahmet",
        "sehir": sehirler[konum],
        "favori_sehirler": sehirler[:3],
        "aktif_mi": True
    },
    "siparis": {
        "tutar": siparis_tutari,
        "kargo_durumu": "50 TL kargo ücreti" if siparis_tutari >= 200 and siparis_tutari < 500 else "Diğer",
        "urunler": urunler,
        "urun_sayisi": len(urunler)
    },
    "stok": [
        {"urun": urunler[0], "adet": 5, "fiyat": 25000},
        {"urun": urunler[1], "adet": adet, "fiyat": 500},
        {"urun": urunler[2], "adet": 7, "fiyat": 1200}
    ],
    "istatistik": {
        "konum": konum,
        "sayac": sayac,
        "silinen_sehir": silinen,
        "toplam_sehir": len(sehirler)
    },
    "ayarlar": {
        "tema": "koyu",
        "bildirimler": ["e-posta", "sms"],
        "diller": ("Türkçe", "English")
    }
}
print(karmasik_veri)

print("\nDict için en çok kullanılan fonksiyonlar ve örnekleri:\n")

# 1) keys()
print("1. keys(): Sözlükteki tüm anahtarları döndürür.")
print("Sonuç:", karmasik_veri.keys(), "\n")

# 2) values()
print("2. values(): Sözlükteki tüm değerleri döndürür.")
print("Sonuç:", list(karmasik_veri.values()), "\n")

# 3) items()
print("3. items(): Sözlükteki anahtar-değer çiftlerini döndürür.")
for anahtar, deger in karmasik_veri.items():
    print(f"  {anahtar}: {deger}")
print()

# 4) get()
print("4. get(key, default): Anahtara karşılık gelen değeri döndürür, yoksa varsayılan değer döner.")
print("'musteri' anahtarı:", karmasik_veri.get("musteri", "Bulunamadı"))
print("'adres' anahtarı:", karmasik_veri.get("adres", "Bulunamadı"), "\n")
karmasik_veri["dataKey"] = {"nestedKey": "nestedValue", "number": 42}

# 5) update()
ornek_dict = karmasik_veri.copy()
ornek_dict.update({"yeni_alan": "yeni_deger", "musteri": {"ad": "Mehmet"}})
print("5. update(dict): Sözlüğü başka bir sözlükle günceller veya yeni anahtar ekler.")
print("Güncellenen 'musteri':", ornek_dict["musteri"])
print("Yeni alan:", ornek_dict["yeni_alan"], "\n")

# 6) pop()
ornek_dict = karmasik_veri.copy()
silinen_deger = ornek_dict.pop("ayarlar", None)
print("6. pop(key, default): Belirtilen anahtarı siler ve değerini döndürür.")
print("Silinen 'ayarlar':", silinen_deger)
print("Kalan anahtarlar:", list(ornek_dict.keys()), "\n")

# 7) popitem()
ornek_dict = karmasik_veri.copy()
son_eleman = ornek_dict.popitem()
print("7. popitem(): Son eklenen anahtar-değer çiftini siler ve döndürür.")
print("Silinen eleman:", son_eleman, "\n")

# 8) setdefault()
ornek_dict = karmasik_veri.copy()
ornek_dict.setdefault("yedek_alan", "varsayılan_deger")
print("8. setdefault(key, default): Anahtar yoksa ekler, varsa mevcut değeri döndürür.")
print("'yedek_alan':", ornek_dict.get("yedek_alan"), "\n")

# 9) copy()
kopya_dict = karmasik_veri.copy()
print("9. copy(): Sözlüğün yüzeysel kopyasını oluşturur.")
print("Aynı nesne mi?:", kopya_dict is karmasik_veri, "\n")

name = "Ali"
name1 = "Ali"

# 10) clear()
ornek_dict = karmasik_veri.copy()
ornek_dict.clear()
print("10. clear(): Sözlükteki tüm elemanları siler.")
print("Temizlenmiş sözlük:", ornek_dict, "\n")

# 11) in operatörü
print("11. 'in' operatörü: Anahtarın sözlükte olup olmadığını kontrol eder.")
print("'musteri' var mı?:", "musteri" in karmasik_veri)
print("'adres' var mı?:", "adres" in karmasik_veri, "\n")

# Ek: len()
print("Ek: len(dict): Sözlükteki anahtar sayısını döndürür.")
print("Anahtar sayısı:", len(karmasik_veri))
```

```python
# Set oluşturma
kategoriler = {"Elektronik", "Giyim", "Elektronik", "Ev", "Giyim"}
print(kategoriler)  # {'Elektronik', 'Giyim', 'Ev'} — tekrarlar gitti

# Set işlemleri
set_a = {"Laptop", "Mouse", "Klavye"}
set_b = {"Laptop", "Monitor", "Webcam"}

print(set_a & set_b)  # Kesişim: {'Laptop'}
print(set_a | set_b)  # Birleşim: {'Laptop','Mouse','Klavye','Monitor','Webcam'}
print(set_a - set_b)  # Fark: {'Mouse', 'Klavye'}


kategoriler_dizi = list(kategoriler)
print(kategoriler_dizi)
```

```python
from typing import Optional

def veri_ozetle(veri: any) -> str:
    if isinstance(veri, dict):
        return f"Sözlük: {len(veri)} anahtar"
    elif isinstance(veri, list):
        return f"Liste: {len(veri)} eleman"
    elif isinstance(veri, set):
        return f"Set: {len(veri)} benzersiz eleman"
    elif isinstance(veri, tuple):
        return f"Tuple: {len(veri)} eleman"
    elif isinstance(veri, str): 
        return f"String: {len(veri)} karakter"
    else:
        return "Bilinmeyen veri türü"
    
print(veri_ozetle("ali"))

# kdv_hesapla fonksiyon kdv_orani: float = 0.18
def kdv_hesapla(fiyat, kdv_orani: Optional[float] = 0.18):
    kdv_tutari = fiyat * kdv_orani
    toplam_fiyat = fiyat + kdv_tutari
    return toplam_fiyat
kdv = kdv_hesapla(100, 0.20)
print(f"KDV dahil fiyat: {kdv:.2f}")


def fnc1(data: str | int) -> str:
    if isinstance(data, str):
        return f"Veri bir string: {data}"
    elif isinstance(data, int):
        return f"Veri bir integer: {data}"
    else:
        return "Bilinmeyen veri türü"
    
def fnc2(data: list | dict) -> tuple[str, int]:
    if isinstance(data, list):
        return f"Veri bir liste: {len(data)} eleman", len(data)
    elif isinstance(data, dict):
        return f"Veri bir sözlük: {len(data)} anahtar", len(data)
    else:
        return "Bilinmeyen veri türü", 0
    
print(fnc2(sehirler)[0]) 
print(fnc2(sehirler)[1]) 
```
